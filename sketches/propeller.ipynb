## Blade properties

Non-dimensional radial coordinate: $\zeta = \dfrac{r}{R}$

Chord-line pitch angle ($\beta_c(\zeta)$): Angle measured from the propeller plane to the section's airfoil
chord.

Chord-line pitch length ($\lambda(\zeta)$): Hypothetical distance the propeller would move forward
in one revolution if it were turned without slipping. Related to the chord-line pitch angle by:

$\lambda_c(\zeta) = 2\pi r \tan(\beta_c) = 2\pi R \zeta \tan(\beta_c)$

Chord-line pitch-to-diameter ratio ($K_c(\zeta)$): Chord-line pitch length normalized with propeller
diameter.

$K_c(\zeta) = \pi \zeta \tan(\beta_c)$

Zero-lift angle of attack ($\alpha_0(\zeta)$): Angle measured from the propeller plane to the section's
airfoil zero-lift line.

Aerodynamic pitch angle ($\beta(\zeta)$): Angle measured from the propeller plane to the section's
airfoil zero-lift line.

$\beta(\zeta) = \beta_c(\zeta) - \alpha_0(\zeta)$ 

Given $K_c(\zeta)$ and $\alpha_0(\zeta)$, we directly obtain $\beta(\zeta)$, which is what we need for
further computations.

## Airflow-related distributions

Total downwash angle ($\varepsilon$): Angle measured from the propeller plane to the local airflow.

Advance angle ($\varepsilon_\infty$): Downwash angle due to the axial propeller aerodynamic
velocity.

Induced angle ($\varepsilon_\infty$): Downwash angle due to the axial propeller aerodynamic velocity.

Angle of attack ($\alpha$): Angle measured from the local airflow to the section's airfoil
zero-lift line.

$\alpha(\zeta) = \beta(\varepsilon) - \varepsilon(\zeta)j$


Aerodynamic angle of attack ($\alpha$): Angle measured from the section's airfoil zero-lift line to
the local airflow.


Specified parameters:
$K_c(\zeta), \alpha_0(\zeta)$

interpolator from Interpolations.jl or a function wrapper around it.
